```python
import pandas as pd
import matplotlib.pyplot as plt

#  CSV 
files = {
    'Grid Search': 'grid_result.csv',
    'Random Search': 'random_result.csv',
    'Bayesian': 'bayesian_result.csv',
    'Genetic Search': 'genetic_result.csv',
    'Successive Halving': 'sh_result.csv'
}

# 
sampling_interval = 50
reasonable_threshold = 90

plt.figure(figsize=(14, 8))

for label, file in files.items():
    data = pd.read_csv("/Users/zyt/Desktop/DL_finalproject/comparison/1/" + file)
    
    #  total_training_time  10000 
    data = data[data['total_training_time'] <= 12500]
    
    #  50 
    sampled_data = data.iloc[::sampling_interval, :]
    line, = plt.plot(sampled_data['total_training_time'], sampled_data['val_accuracy'], label=label, linestyle='-', alpha=0.7)
    color = line.get_color()

    # 
    best_model = data.loc[data['val_accuracy'].idxmax()]
    plt.scatter(best_model['total_training_time'], best_model['val_accuracy'], color=color, s=100, marker='*', label=f'{label} Best Model')
    
    # 
    reasonable_model = data[data['val_accuracy'] >= reasonable_threshold].iloc[0] if not data[data['val_accuracy'] >= reasonable_threshold].empty else None
    if reasonable_model is not None:
        plt.scatter(reasonable_model['total_training_time'], reasonable_model['val_accuracy'], color=color, s=80, marker='o', label=f'{label} Reasonable Model')

    # Genetic Search  generation 
    if label == 'Genetic Search':
        if 'generation' in data.columns:
            generations = data['generation'].unique()
            gen_points = [data[data['generation'] == gen].iloc[-1]['total_training_time'] for gen in generations]
            
            #  generation 
            for i, x in enumerate(gen_points):
                plt.text(x, -0.1, f'Gen {i+1}', ha='center', color=color, fontsize=8, transform=plt.gca().get_xaxis_transform())

    # Successive Halving  stage 
    if label == 'Successive Halving':
        if 'stage' in data.columns:
            stages = data['stage'].unique()
            stage_points = [data[data['stage'] == stage].iloc[-1]['total_training_time'] for stage in stages]
            
            #  stage 
            for i, x in enumerate(stage_points):
                plt.text(x, -0.15, f'Stage {i+1}', ha='center', color=color, fontsize=8, transform=plt.gca().get_xaxis_transform())

# 
plt.xlabel("Total Training Time")
plt.ylabel("Accuracy")
plt.title("Training Trend Comparison with Total Training Time <= 12500 Across Algorithms")
plt.legend(loc="lower right")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# 
folders = ['1', '2', '3']
base_path = "/Users/zyt/Desktop/DL_finalproject/comparison/"
files = {
    'Grid Search': 'grid_result.csv',
    'Random Search': 'random_result.csv',
    'Bayesian': 'bayesian_result.csv',
    'Genetic': 'genetic_result.csv',
    'Successive Halving (10 Epochs)': 'sh_result.csv',
    'Successive Halving (Global)': 'sh_result.csv'
}

# 
results = {
    'Algorithm': [],
    'Experiment': [],
    'Metric': [],
    'Accuracy': []
}

# 
for folder in folders:
    for algo, file_name in files.items():
        file_path = os.path.join(base_path, folder, file_name)
        
        if os.path.exists(file_path):
            data = pd.read_csv(file_path)
            
            # Successive Halving 10 Epochs  Global
            if 'Successive Halving' in algo:
                if '10 Epochs' in algo:
                    subset_data = data[data['epochs'] == 10]  #  epoch 10 
                else:
                    subset_data = data  # 
            else:
                subset_data = data
            
            # 
            best_model = subset_data.loc[subset_data['val_accuracy'].idxmax()]
            results['Algorithm'].append(algo)
            results['Experiment'].append(f"Experiment {folder}")
            results['Metric'].append('Train Accuracy')
            results['Accuracy'].append(best_model['train_accuracy'])
            
            results['Algorithm'].append(algo)
            results['Experiment'].append(f"Experiment {folder}")
            results['Metric'].append('Val Accuracy')
            results['Accuracy'].append(best_model['val_accuracy'])
        else:
            #  None 
            for metric in ['Train Accuracy', 'Val Accuracy']:
                results['Algorithm'].append(algo)
                results['Experiment'].append(f"Experiment {folder}")
                results['Metric'].append(metric)
                results['Accuracy'].append(None)

#  DataFrame
results_df = pd.DataFrame(results)

# 
plt.figure(figsize=(16, 10))
sns.set(style="whitegrid")

# 
colors = {'Train Accuracy': 'skyblue', 'Val Accuracy': 'salmon'}

# 
algorithms = results_df['Algorithm'].unique()
experiments = [f"Experiment {folder}" for folder in folders]

# 
bar_width = 0.2  # 
space_between_algorithms = 1  # 
for idx, algo in enumerate(algorithms):
    # 
    algo_data = results_df[results_df['Algorithm'] == algo]
    for exp_idx, experiment in enumerate(experiments):
        exp_data = algo_data[algo_data['Experiment'] == experiment]
        
        for metric_idx, metric in enumerate(['Train Accuracy', 'Val Accuracy']):
            metric_data = exp_data[exp_data['Metric'] == metric]['Accuracy'].values
            bar_position = (idx * (len(experiments) + space_between_algorithms) + exp_idx) + metric_idx * bar_width
            plt.bar(
                bar_position, metric_data[0], width=bar_width, color=colors[metric],
                label=f'{metric}' if idx == 0 and exp_idx == 0 else ""
            )
            
            # 
            plt.text(
                bar_position, metric_data[0] + 0.01, f'{metric_data[0]:.2f}',
                ha='center', va='bottom', fontsize=8
            )

#  x 
xtick_positions = [(i * (len(experiments) + space_between_algorithms)) + (len(experiments) / 2) - 0.5 for i in range(len(algorithms))]
plt.xticks(xtick_positions, algorithms, rotation=45)
plt.xlabel("Algorithm")
plt.ylabel("Accuracy")
plt.title("Comparison of Train and Validation Accuracies Across Experiments")
plt.legend(loc="upper left", title="Metric")
plt.tight_layout()
plt.show()
```
